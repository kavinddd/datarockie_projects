```python
import requests
import pandas as pd
```

### Pokemon API
<br>The purpose of this mini project is just to pratice using public API by python. </br>
<br>https://pokeapi.co/</br>

```python
main = {'id':[], 'name':[], 'hp':[],'atk':[], 'def':[], 'sp_atk':[], 'sp_def':[], 'speed':[]}
stats_col = ['hp','atk','def','sp_atk','sp_def','speed']
pk_types = dict()
# 30 pokemons
for i in range(1,31):
    # Request data of each pokemon
    url = "https://pokeapi.co/api/v2/pokemon/"+str(i)
    response =requests.get(url)
    result = response.json()
    pk_id = result['id']
    
    # Storing data
    main['id'].append(pk_id)
    main['name'].append(result['name'])
    for i, stat in enumerate(stats_col):
        main[stat].append(result['stats'][i]['base_stat'])
    pk_types[pk_id] = dict()
    pk_types[pk_id]['types'] = [pk_type['type']['name'] for pk_type in result['types']]
print('Done')
```

```python
# Prepare data to build right dataframe
# split pokemon types into 3 types of columns
dict1 = {'type_1':[], 'type_2':[]}
for value in pk_types.values():
    for i in range(2):
        try:
            n_type = f"type_{i+1}"
            dict1[n_type].append(value['types'][i])
        except IndexError:
            dict1[n_type].append('-')
print(dict1)
```

```python
left_df = pd.DataFrame(main)
right_df = pd.DataFrame(dict1, index=pk_types.keys())
# Create total stats before merging two columns together
left_df['total'] = left_df[stats_col].sum(axis=1)
print(left_df.head())
print(right_df.head())
```

```python
# merge two data frame together
full_df = left_df.merge(right_df, left_on='id', right_on=right_df.index)
full_df
```
